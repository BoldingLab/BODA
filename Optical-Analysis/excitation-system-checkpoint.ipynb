```python
%matplotlib inline

from rayoptics.environment import *

# Create a new optical model
opm = OpticalModel()

sm = opm['seq_model']
osp = opm['optical_spec']
pm = opm['parax_model']
em = opm['ele_model']
pt = opm['part_tree']

sm.gaps[0].thi = 0.8 #Working distance

sm.do_apertures = False # Disable automatic aperture calculation

opm.add_from_file("./lens-data/CODV_45078.seq", t=0.3) #6.0mm Dia. x 6.0mm FL, MgF2 Coated, Plano-Convex Lens

opm.add_from_file("./lens-data/CODV_45078.seq", t=8.6) #6.0mm Dia. x 6.0mm FL, MgF2 Coated, Plano-Convex Lens

opm.add_from_file("./lens-data/CODV_32299.seq", t=0.3) #6.25mm Dia. x 12.5mm FL, MgF2 Coated, Achromatic Doublet Lens

opm.add_from_file("./lens-data/CODV_45345.seq", t=4.4) #6.25mm Dia. x 60mm FL, MgF2 Coated, Achromatic Doublet Lens

e1 = pt.obj_by_name('E1')
opm.flip(e1)

e2 = pt.obj_by_name('E2')
opm.flip(e2)

ce1 = pt.obj_by_name('CE1')

ce2 = pt.obj_by_name('CE2')
opm.flip(ce2)


opm.update_model()

em.list_model()
sm.list_model()
pt.list_model()

# Visualize the complete system
layout_plt = plt.figure(FigureClass=InteractiveLayout, opt_model=opm,
                        do_draw_rays=True, do_paraxial_layout=False).plot()
```
