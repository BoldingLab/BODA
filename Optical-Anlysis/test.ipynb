```python
%matplotlib inline

from rayoptics.environment import *

# Create a new optical model
opm = OpticalModel()
sm = opm['seq_model']
osp = opm['optical_spec']
pm = opm['parax_model']
em = opm['ele_model']
pt = opm['part_tree']

sm.gaps[0].thi = 5 #Working distance

# Disable automatic aperture calculation
sm.do_apertures = False

# Set the entrance pupil diameter
#osp.pupil.value = 22

opm.add_from_file("./lens-data/CODV_45345.seq", t=20.0) #6.25mm Dia. x 60mm FL, MgF2 Coated, Achromatic Doublet Lens

opm.add_from_file("./lens-data/CODV_45078.seq", t=20.0) #6.0mm Dia. x 6.0mm FL, MgF2 Coated, Plano-Convex Lens

opm.add_from_file("./lens-data/CODV_32299.seq", t=20.0) #6.25mm Dia. x 12.5mm FL, MgF2 Coated, Achromatic Doublet Lens

opm.add_from_file("./lens-data/CODV_32301.seq", t=20.0) #6.25mm Dia. x 15mm FL, MgF2 Coated, Achromatic Doublet Lens


opm.update_model()


# Visualize the complete system
layout_plt = plt.figure(FigureClass=InteractiveLayout, opt_model=opm,
                        do_draw_rays=True, do_paraxial_layout=False).plot()
```
